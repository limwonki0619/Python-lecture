- pinkwink web 의 hit 수

```python
pinkwink_web = pd.read_csv('data/07. PinkWink Web Traffic.csv', 
                                          encoding='utf-8', thousands=',',
                                          names = ['date','hit'], index_col=0)
pinkwink_web = pinkwink_web[pinkwink_web['hit'].notnull()]
pinkwink_web.head()
```

```python
pinkwink_web['hit'].plot(figsize=(12,4), grid=True, color='coral')
plt.show()
```

```python
time = np.arange(0,len(pinkwink_web))
traffic = pinkwink_web['hit'].values

fx = np.linspace(0, time[-1], 1000)  # 1000 등분 

print(time[:10])
print(traffic[:10])
```

```python
def mse(f, x, y):
    return np.sqrt(np.mean((f(x)-y)**2))  # MSE : Mean Square Error
```

- polyfit ( 선형 회귀식을 찾아준다. 즉, 기울기와 절편을 구해줌)
- polyfit (data1, data2, diff)

```python
fp1 = np.polyfit(time, traffic, 1)  # 1차식
f1 = np.poly1d(fp1)

f2p = np.polyfit(time, traffic, 2)  # 2차식 
f2 = np.poly1d(f2p)

f3p = np.polyfit(time, traffic, 3)  # 3차식
f3 = np.poly1d(f3p)

f15p = np.polyfit(time, traffic, 15) # 15차식
f15 = np.poly1d(f15p)

print(mse(f1, time, traffic))
print(mse(f2, time, traffic))
print(mse(f3, time, traffic))
print(mse(f15, time, traffic))  # MSE가 차수가 높아질수록 낮아진다. 즉, 오자체곱합이 감소한다. 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time, traffic, s=10)  # 산점도, S = marker size 

plt.plot(fx, f1(fx), lw=4, label='f1')
plt.plot(fx, f2(fx), lw=4, label='f2')
plt.plot(fx, f3(fx), lw=4, label='f3')
plt.plot(fx, f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')

plt.legend(loc=2)  # loc : location 
plt.show()
```

## **2. Prophet 모듈을 이용한 forecast 예측**

```python
df = pd.DataFrame({'ds':pinkwink_web.index, 'y':pinkwink_web['hit']})
df.reset_index(inplace=True)
df['ds'] =  pd.to_datetime(df['ds'], format="%y. %m. %d.")
del df['date']

m = Prophet(yearly_seasonality=True, daily_seasonality=True)  # 연 단위와 일 단위 hit 수에 주기성을 가지는가?, 
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=60)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
plt.show()
```

```python
m.plot_components(forecast)
plt.show()

# plot 1. 연 단위 
# plot 2. 주간 단위
# plot 3. 월 단위
# plot 4. 시간 단위
```
