## 19대 대선 개표 결과 데이터 획득하기

### 광역시도 및 시군구 데이터 확보하기

```python
sigun_names_value[:5]
```

```python
from tqdm import tqdm_notebook

sido_name_list = []
sigun_name_list = []

for sido_value in tqdm_notebook(sido_names_value):
    element = driver.find_element_by_css_selector('#cityCode') # 시도 선택창 위치
    element.send_keys(sido_value)  # 시도 선택창에 각 시도 선택하기
    
    time.sleep(1)
    
    sigun_list_raw = driver.find_elements_by_css_selector('#townCode > option')  # 각 시도별 시군구 정보 얻기
    sigun_names_values = [option.text for option in sigun_list_raw]
    
    sigun_names_values = sigun_names_values[1:] # 불필요 요소 삭제 
    
    for sigun_value in sigun_names_values:
        sido_name_list.append(sido_value) # 시군구 개수와 맞추기위한 반복
        sigun_name_list.append(sigun_value) # 시군구 정보 
```

```python
election_result = pd.DataFrame(
    {'시도명':sido_name_list,'시군구명':sigun_name_list})
election_result.tail()
```

### 시군구별 득표수 데이터 확보하기

```python
# 투표인 수와 문, 홍, 안 후보의 특표수를 저장하는 함수 
def get_vote_info(n):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    tmp = soup.find_all('td', 'alignR')[1:5]
    tmp_values = []
    
    for tmp_val in tmp:
        tmp_values.append(float(tmp_val.get_text().replace(',', '')))
        
    pop[n] = tmp_values[0]
    moon[n] = tmp_values[1]
    hong[n] = tmp_values[2]
    ahn[n] = tmp_values[3]
```

```python
# 에러 발생시 실행할 함수 

import numpy as np

def fail_procedure(n):
    pop[n]  = np.nan
    moon[n] = np.nan
    hong[n] = np.nan
    ahn[n]  = np.nan
```

```python
# 최종결과가 저장될 리스트 생성

pop = [np.nan]*len(election_result)
moon = [np.nan]*len(election_result)
hong = [np.nan]*len(election_result)
ahn = [np.nan]*len(election_result)

len(pop), len(moon), len(hong), len(ahn)
```

### 득표수 데이터 획득

```python
for n in tqdm_notebook(election_result.index):
    try:
        sido = driver.find_element_by_css_selector('#cityCode')
        sido.send_keys(election_result['시도명'][n])
        time.sleep(0.5)
        
        sigun = driver.find_element_by_css_selector('#townCode')
        sigun.send_keys(election_result['시군구명'][n])
        time.sleep(0.5)
        
        driver.find_element_by_css_selector('#searchBtn').click()
        time.sleep(0.5)
        
        get_vote_info(n)
        time.sleep(0.5)
        
    except:
        print('Error', n)
        fail_procedure(n)
```

```python
election_result['pop'] = pop
election_result['moon'] = moon
election_result['hong'] = hong
election_result['ahn'] = ahn
```

```python
# nan이 있는 지역만 다시 정보 요청하기 
# 웹데이터의 경우 에러가 안나는 경우는 드물다..
# 결측치가 없어질 때 까지 반복해서 실행한다. 

re_try_index = election_result[election_result['pop'].isnull()].index

for n in tqdm_notebook(re_try_index):
    try:
        sido = driver.find_element_by_css_selector('#cityCode')
        sido.send_keys(election_result['시도명'][n])
        time.sleep(0.5)
        
        sigun = driver.find_element_by_css_selector('#townCode')
        time.sleep(0.5)
        
        sigun.send_keys(election_result['시군구명'][n])
        time.sleep(0.5)
        
        driver.find_element_by_css_selector('#searchBtn').click()
        time.sleep(0.5)
        
        get_vote_info(n)
        time.sleep(0.5)
        
    except:
        print('Error', n)
        fail_procedure(n)
        time.sleep(0.5)
```

```python
election_result['pop'] = pop
election_result['moon'] = moon
election_result['hong'] = hong
election_result['ahn'] = ahn
```

### 각 후보의 득표율과 지역 ID 정리하기

#### 지역별 ID 생성 및 병합

```python
election_result = pd.read_csv('election_result.csv', encoding='CP949')
election_result.head()
```

```python
def GetNoneList(data):
    """
    데이터프레임의 행 갯수만큼 빈 리스트를 만드는 함수
    """
    global si_name  # 전역변수 설정
    si_name = [None] * len(data)
```

```python
GetNoneList(election_result)
```

```python
tmp_gu_dict = {'수원':['장안구', '권선구', '팔달구', '영통구'], 
               '성남':['수정구', '중원구', '분당구'], 
               '안양':['만안구', '동안구'], 
               '안산':['상록구', '단원구'], 
               '고양':['덕양구', '일산동구', '일산서구'], 
               '용인':['처인구', '기흥구', '수지구'], 
               '청주':['상당구', '서원구', '흥덕구', '청원구'], 
               '천안':['동남구', '서북구'], 
               '전주':['완산구', '덕진구'], 
               '포항':['남구', '북구'], 
               '창원':['의창구', '성산구', '진해구', '마산합포구', '마산회원구'], 
               '부천':['오정구', '원미구', '소사구']}
```

```python
# 데이터가 tmp_gu_dict에 있는 시군구명이 띄어쓰기가 안되있을 경우 아래의 코드 실행 
# 만약 띄어쓰기가 되어있는 경우 5장에있는 GetID함수를 사용 
# 기존 GetID와 함수가 약간 다름 

def GetID(df):
    for n in df.index:
        if df['시도명'][n][-3:] not in ['광역시', '특별시', '자치시']:  # 광역시도 열에서 뒷 세자리가 '광역시', '특별시', '자치시' 가 아니면,
        
            # 같은 '시도' 이름을 가지는 고성 지역 처리
            if df['시군구명'][n].strip()[:-1]=='고성' and df['시도명'][n]=='강원도':  # '시도'가 고성이면서, '광역시도'가 강원도인 지역 처리
                si_name[n] = '고성(강원)'
            elif df['시군구명'][n].strip()[:-1]=='고성' and df['시도명'][n]=='경상남도':  # '시도'가 고성이면서, '광역시도가' 경남인 지역 처리
                si_name[n] = '고성(경남)'
            else:
                si_name[n] = df['시군구명'][n].strip()[:-1]  # 나머지 지역은 맨 뒤의 '구' 삭제 
        
            # 광역시가 아니면서 구를 가지고 있는 시 처리 
            for keys, values in tmp_gu_dict.items():
                if len(df['시군구명'][n]) >= 5 and df['시군구명'][n].strip()[:2] in keys:
                    if len(df['시군구명'][n].split('시')[1]) == 2: # 수원시장안구 처럼 띄어쓰기가 안되어있을 경우 '시'를 기준으로 나눈다.
                        si_name[n] = keys + ' ' + df['시군구명'][n].split('시')[1]
                    elif df['시군구명'][n].split('시')[1] in ['마산합포구','마산회원구']:  # 마산합포와 회원구는 '합포', '회원'만 출력
                        si_name[n] = keys + ' ' + df['시군구명'][n].strip()[5:-1]
                    else:
                        si_name[n] = keys + ' ' + df['시군구명'][n].strip()[3:-1]  # '구'이름이 3글자 이상인 경우 맨 뒤의 '구' 삭제  
        
        elif df['시도명'][n] == '세종특별자치시':  # 세종자치시의 경우는 세종으로
            si_name[n] = '세종'
        
        else:  # tmp_gi_dict와 세종을 제외한 나머지 지역 처리 
            if len(df['시군구명'][n].strip()) >= 3:
                si_name[n] = df['시도명'][n][:2] + ' ' + df['시군구명'][n].strip()[:-1]
            else:
                si_name[n] = df['시도명'][n][:2] + ' ' + df['시군구명'][n].strip()
        
        print(si_name[n], n)
```

```python
# GetID(election_result) # 실행결과가 너무 길어 주석처리
```

```python
election_result['ID'] = si_name  # election_result에 ID 컬럼 생성 
```

#### 각 후보의 득표율 계산

```python
election_result[['rate_moon', 'rate_hong', 'rate_ahn']] = \
    election_result[['moon', 'hong', 'ahn']].div(election_result['pop'], axis=0) * 100

election_result.head()
```

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] < whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

```python
# 문재인 vs 홍준표 득표율 
drawKorea('hong_vs_moon', final_data, 'RdBu') # 문재인이 크게 이긴경우 진한 파랑, 홍준표가 크게 이긴경우 진한 빨강 
```
