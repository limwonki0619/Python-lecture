## **04.01 Pandas 패키지의 소개 : 연습 문제**

### **연습 문제 1** 

- 임의로 두 개의 시리즈 객체를 만든다.  
- 모두 문자열 인덱스를 가져야 하며 두 시리즈에 공통적으로 포함되지 않는 라벨이 있어야 한다.
- 위에서 만든 두 시리즈 객체를 이용하여 사칙 연산을 한다.

```python
import numpy as np
import pandas as pd
```

```python
Series_1 = pd.Series(list(range(4)), # 첫번째 시리즈 
                    index = ['서울', '부산', '인천', '대구'])
Series_1
```

```python
Series_2 = pd.Series(list(range(9, 5, -1)), # 두번째 시리즈
                    index = ['서울', '부산', '대구', '울산'])
Series_2
```

```python
# 사칠연산 1 : 더하기 
Series_1 + Series_2
```

```python
# 사칙연산 2 : 뺴기
Series_1 - Series_2
```

```python
# 사칙연산 3 : 곱하기
Series_1 * Series_2
```

```python
# 사칙연산 4 : 나누기 
Series_1 / Series_2
```

### **연습 문제 2**

다음 조건을 반족하는 임의의 데이터프레임을 하나 만든다. 
- 열의 갯수와 행의 갯수가 각각 5개 이상이어야 한다.
- 열에는 정수, 문자열, 실수 자료형 데이터가 각각 1개 이상씩 포함되어 있어야 한다.

```python
p2_data = {
    '이름' : ['강모씨', '김모씨', '장모씨', '이모씨', '박모씨'],
    '나이' : [27, 28, 29, 30, 25],
    '지역' : ['대구', '대전', '대구', '대구', '대전'],
    '키'   : [172.4, 180.3, 182.1, 175.8, 179.5],
    '차'   : ['아반뗴', '소나타', 'G70', '벨로스터', 'K3']
}


p2_columns = list(p2_data.keys())
p2_index = list(range(1, 6))

p2_df = pd.DataFrame(p2_data, index = p2_index, columns = p2_columns)
p2_df
```

### **연습 문제 3**

다음 데이터프레임에서 지정하는 데이터를 뽑아내거나 처리하라.

```python
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
# 1. 모든 학생의 수학점수를 시리즈로 나타낸다.

df['수학']
```

```python
# 2. 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.

df[['국어', '영어']]
```

```python
# 3. 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.

df['평균점수'] = np.round(np.mean(df, axis=1), 2)
df
```

```python
# 4. 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산하라.
df2 = df.copy()  # copy없이 실행하면 경고 메세지가 발생(?)

df2['영어']['방자'] = 80
df2['평균점수'] = np.round(np.mean(df2, axis=1), 2)
df2
```

## **04.03 데이터프레임 고급 인덱싱 : 연습 문제**

### **연습 문제 1**

1. 모든 행과 열에 라벨을 가지는 5 x 5 이상의 크기를 가지는 데이터프레임을 만든다.
2. 10가지 이상의 방법으로 특정한 행과 열을 선택한다.

```python
np.random.seed(10)
ad_df = pd.DataFrame(np.random.randint(25, size=(5,5)), columns=['A', 'B', 'C', 'D', 'E'], 
                     index=['가', '나', '다', '라', '마'])
ad_df
```

```python
# 1. '가'행의 값 출력 
ad_df.loc['가']
```

```python
# 2. '가'행 ~ '다'행 출력 
ad_df.loc['가':'다']
```

```python
# 3. '나'행 이후의 모든 행 이면서 'C'열에 해당하는 값 출력 
ad_df.loc['나':, 'C']  
```

```python
# 4. '가','다' 행에 해당하는 값 출력 
ad_df.loc[['가', '다'], :]
```

```python
# 5. '다', '마' 행이면서, 'C'열 이후의 값 출력 
ad_df.loc[['다', '마'], 'C':]
```

```python
# 6. '나' ~ '다' 행이면서, 'B','C'열에 해당하는 값 출력 
ad_df.loc['나':'라','B':'C']
```

```python
# 7. 'B' 열에서 값이 10이상인 행과, 'B','D'열에 해당하는 값 출력 
ad_df.loc[ad_df.B > 10, ['B','D']]
```

```python
ad_df # 위로 올라가기 번거러워 다시 출력 
```

```python
# 8. '가' 행의 값을 출력 (iloc)
ad_df.iloc[1]
```

```python
# 9. '다' ~ '마'행과 'C' ~ 'E'열에 해당하는 값 출력 (iloc)
ad_df.iloc[2:,2:]  # 슬라이싱 주의
```

```python
# 10. '가', '다','마' 행이면서, 'B','D'열에 해당하는 값 출력 (iloc)
ad_df.iloc[0:5:2, 1::2]
```

### **연습 문제 3**

타이타닉호 승객을 사망자와 생존자 그룹으로 나누고, 각 그룹에 대해 '미성년자','청년','중년','장년','노년'승객의 비율을 구한다. 각 그룹 별로 비율의 전체 합은 1이 되어야 한다.

bins = [1, 15, 25, 35, 60, 99]  
labels = ["미성년자", "청년", "중년", "장년", "노년"]

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic
```

#### **1. 사망자 그룹**

```python
aliveNo_titanic = titanic.loc[titanic['alive'] == 'no', ['age','alive']]
aliveNo_titanic
```

```python
aliveNo_titanic['age_cut'] = pd.cut(aliveNo_titanic.age, bins, labels = labels)
aliveNo_titanic
```

```python
aliveNo = aliveNo_titanic['age_cut'].value_counts() / aliveNo_titanic['age_cut'].value_counts().sum()
print(aliveNo)
print(sum(aliveNo.values))
```

### **2. 생존자 그룹**

```python
aliveYes_titanic = titanic.loc[titanic['alive'] == 'yes', ['age','alive']]
aliveYes_titanic
```

```python
aliveYes_titanic['age_cut'] = pd.cut(aliveYes_titanic.age, bins, labels = labels)
aliveYes_titanic
```

```python
aliveYes = aliveYes_titanic['age_cut'].value_counts() / aliveYes_titanic['age_cut'].value_counts().sum()
print(aliveYes)
print(sum(aliveYes.values))
```

### **생존자/사망자 그룹 간 나이별 생존비율 비교**

```python
print(aliveNo.sort_index())  # 사망자 그룹 
print(aliveYes.sort_index()) # 생존자 그룹 
```
