## **연습 문제 6-1**

두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.

1. 각각 5 x 5 이상의 크기를 가진다.
2. 공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다.

```python
import pandas as pd
data1 = {
    '차량코드' : [1001, 1002, 1003, 1004, 1005],
    '차량명' : ['베뉴', '코나', '소나타', '아반떼', '셀토스'],
    '제조사' : ['현대차', '현대차', '현대차', '현대차', '기아차'],
    '가격' : [2000, 2800, 3200, 2300, 3000],
    '차급' : ['소형', '소형', '중형', '준중형', '소형']
}

columns1 = data1.keys()
df1 = pd.DataFrame(data1, columns=columns1)

df1
```

```python
data2 = {
    '코드' : [1001, 1003, 1004, 1006, 1007],
    '중고가격' : ['정보없음', '정보없음', '정보없음', '정보없음', '정보없음'],
    '최고연비' : [13.3, 9.8, 10.6, 7.3, 8.4],
    '최저연비' : [13.7, 13.3, 17.8,  14.6, 15.0],
    '연료' : ['가솔린', '가솔린', '가솔린', '가솔린', '가솔린']
         }
columns2 = data2.keys()

df2 = pd.DataFrame(data2, columns = columns2)
df2
```

```python
df3 = pd.merge(df1, df2, how='left', left_on='차량코드', right_on='코드').drop('코드', axis=1)  # '코드'열 삭제 
df3
```

## **연습 문제 6-2**

어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 실적 정보는 "매출", "비용", "이익" 으로 이루어진다. (이익 = 매출 - 비용).

또한 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
import numpy as np
import pandas as pd

# 전반기 데이터프레임 
p_1_6 = {
    '매출' : [5000, 5500, 6600, 5000, 3000, 2500],
    '비용' : [2000, 1500, 2100, 3000, 3000, 2800],
}

p1_col = p_1_6.keys()
p1_index = ['1월', '2월', '3월', '4월', '5월', '6월']

df_p1 = pd.DataFrame(p_1_6, columns = p1_col, index = p1_index)
df_p1
```

```python
# 후반기 데이터프레임 

p_7_12 = {
    '매출' : [2000, 1500, 2000, 3000, 6000, 8000],
    '비용' : [3000, 3000, 2500, 3000, 4000, 4000]
}

p2_col = p_1_6.keys()
p2_index = ['7월', '8월', '9월', '10월', '11월', '12월']

df_p2 = pd.DataFrame(p_7_12, columns = p2_col, index = p2_index)
df_p2
```

```python
p_all = pd.concat([df_p1, df_p2], axis = 0)  # 열 방향 데이터 연결 
p_all
```

```python
# '이익' 컬럼 생성 
p_all['이익'] = p_all['매출'] - p_all['비용']
p_all
```

```python
# '촐 실적' 인덱스 생성 

p_all.loc['총 실적', :] = p_all.sum(axis=0)  # 열 방향 더하기 (디폴트)
p_all
```

## **연습 문제 7-1**

key1의 값을 기준으로 data1의 값을 분류하여 합계를 구한 결과를 시리즈가 아닌 데이터프레임으로 나타내라.

```python
import numpy as np
import pandas as pd

np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
to_df = pd.DataFrame(df2.data1.groupby(df2.key1).sum())  # key1 별 합계
print(type(to_df))
to_df
```

## **연습문제 7-3**

팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴본다.
어떤 요인이 가장 크게 작용하는지 판단할 수 있는 방법이 있는가

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(2)  # 팁의 비율 
tips.tail() # 기본 데이터 생성 및 확인
```

```python
# 데이터 : ttip_pct
# 인덱스(행) : size
# 컬럼(열) : day, time 

tips.pivot_table('tip_pct', 'size', ['day', 'time'])  
```

## **연습 문제 7-4**

타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.

- titanic = sns.load_dataset("titanic")
- qcut 명령으로 세 개의 나이 그룹을 만든다.
- 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.
- 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# 0 전체 데이터 갯수 확인 

titanic.count()
```

```python
# 1. qcut 명령으로 세 개의 나이 그룹을 만들어라.

titanic['age_groups'] = pd.qcut(titanic.age, 3, labels=['age1', 'age2', 'age3'])
titanic.age_groups.head(5)
```

```python
# 2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
# 단, 행에는 성별 및 나이그룹에 대한 다중인덱스를 사용하고
# 열에는 선실 인덱스를 사용한다. 

# 방법 1. 우선, 변수 각각의 생존율을 계산해본 후 groupby 메서드를 이용해서 문제를 풀어본다.

# 1.1 성별 생존율 
titanic.groupby('sex')[['survived']].count() / titanic.age_groups.count()  
```

```python
# 1.2 선실별 생존율 

titanic.groupby('class')[['survived']].count() / titanic.age_groups.count()   
```

```python
# 1.3 나이그룹별 생존율 

titanic.groupby('age_groups')[['survived']].count() / titanic.age_groups.count()  
```

```python
# 1.4 성 및 나이그룹별(인덱스), 선실별(컬럼) 생존율 

# survived : 데이터
# mean() : pivot_table의 경우 디폴트 계산값이 평균이지만, groupby 메서드를 사용할 경우 mean() 메서드가 필요함. 
# groupby : sex(행1), age_groups(행2), class(행3)
# unstack으로 행3을 열1로 변환 

titanic['survived'].groupby([titanic["sex"], titanic['age_groups'], titanic['class']]).mean().unstack("class")
```

```python
# 방법 2. pivot_table을 이용한 방법 

# 2.1 전체 인원의 그룹별 인원 
titanic_total_num = titanic.pivot_table('survived', ['sex', 'age_groups'], 'class', aggfunc='count')
titanic_total_num
```

```python
# 2.2 생존자들의 그룹별 인원 
Alive = titanic[titanic.alive == 'yes']

alive_total_count = Alive.pivot_table('survived', ['sex', 'age_groups'], 'class', aggfunc='count')
alive_total_count
```

```python
# 2.3 그룹별 생존율 계산 결과 
alive_total_count / titanic_total_num 
```

```python
# pivot_table을 이용해 한번에 해보기 

# survived : 데이터
# groupby : sex(행1), age_groups(행2), class(행3)
# unstack으로 행3을 열1로 변환 
prob2 = titanic.pivot_table('survived', ['sex', 'age_groups'], 'class')  # 앞에서 차례로 데이터, 인덱스1, 인덱스2, 컬럼1
print(type(prob2))
prob2
```

```python
# 3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만들어라. 

prob3 = titanic.pivot_table('survived', 'sex', 'class')  # 데이터, 행1, 열1, 피봇테이블은 디폴트 계산값이 평균이다.
print(type(prob3))
prob3
```
