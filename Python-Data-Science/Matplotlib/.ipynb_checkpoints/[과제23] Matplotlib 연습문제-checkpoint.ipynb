## **연습문제 1**

여러가지 함수를 사용하여 아래 조건에 맞는 그래프를 그린다. 

 - xlabel, ylabel, title을 모두 갖추고 있어야 한다.
 - 하나의 Figure(일단, 그림이라고 이해한다. 아래에 자세한 설명이 있다.)에 3개 이상의 Plot을 그린다. 
 - 각 Plot은 다른 선, 마크, 색 스타일을 가진다. 
 - legend는 그래프와 겹치지 않는 곳에 위치 시키도록 한다. 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats

font_fname = 'C:/Windows/Fonts/NanumBarunGothic.ttf'  # NanumBarunGothic.ttf를 원하는 폰트명으로 바꾼다.
font_family = fm.FontProperties(fname=font_fname).get_name()  # 해당 폰트명에서 폰트 패밀리 이름(영문)을 가져온다.
print(font_family)                         # 설정한 폰트의 폰트 패밀리명 확인

# 주피터노트북의 경우, 매직(magic) 명령으로 노트북 내부에 그림을 표시하도록 지정
%matplotlib inline 
```

```python
plt.rcParams['font.family'] = font_family  # 폰트패밀리 이름을 전역 폰트로 설정해준다.
plt.rcParams["font.size"] = 20             # 전역 폰트 사이즈 
plt.rcParams["figure.figsize"] = (16,9)    # pigure size 설정 (가로, 세로)

mpl.rc('axes', unicode_minus=False) 

# -------------------------------------------- 
def Normal_create(start, end, interval, mu, sd, custom):
    xx = np.linspace(start, end, interval)
    rv = sp.stats.norm(mu, sd)
    pdf = rv.pdf(xx)
    plt.plot(xx, pdf, custom)
    plt.title("mu = {0} sd = {1} 정규분포 확률밀도함수".format(mu, sd))
    plt.xlabel("$x$")
    plt.ylabel("$p(x)$")
    plt.ylim(0,1)
    plt.show()
    
# plot1    
Normal_create(-3, 3, 600, 0, 0.45, 'r--')

# plot2
Normal_create(-3, 3, 600, 0, 0.75, 'yo-.')

# plot3
Normal_create(-3, 3, 600, 0, 1, 'b2--')

```
